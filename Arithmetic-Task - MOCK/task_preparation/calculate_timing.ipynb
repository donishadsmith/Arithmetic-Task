```python
columns = df.columns[8:14]; avg_trial_time = []; total_run_time = []; curr_range = [float("inf"), float("-inf")]

for iteration, file in enumerate(files):
    df = pd.read_csv(file, sep = "\t")
    timing_list = np.array([df.loc[index, columns].values.sum() for index in df.index])
    total_run_time.append(timing_list.sum()); avg_trial_time.append(total_run_time[iteration]/total_stim)
    curr_trial_range = [min(timing_list.min(), curr_range[0]), max(timing_list.max(), curr_range[1])]
    
    if file == files[-1]: print(f'Avg trial time (sec): {np.array(avg_trial_time).sum()/len(files)}\nTrial time range: {curr_trial_range}\nAvg run time (min): {np.array(total_run_time).sum()/len(files)*1/60}\nRun time range: {[min(total_run_time)/60, max(total_run_time)/60]}')
```
