```python
import os, pandas as pd, numpy as np, shutil; from glob import glob

#Assign work_dir
work_dir = os.path.join(os.path.dirname(os.getcwd()),"config"); os.chdir(work_dir)
#Gather config files
files = glob("./*.tsv"); length = [pd.read_csv(file, sep = "\t").shape[0] for file in files]
# Make new directories
os.makedirs("english"); os.makedirs("spanish")
#Copy and move fils
[shutil.copy(file,".\\english" + file) for file in files]
[os.rename(file, ".\\spanish" + file) for file in files]
```

```python
#Replace word trials with english_word
for file in glob(".\\english\\*.tsv"):
    df = pd.read_csv(file, sep="\t")
    df.equation_representation = [x.replace("word", "english_word") if x == "word" else x for x in df.equation_representation]
    df.comparison_representation = [x.replace("word", "english_word") if x == "word" else x for x in df.comparison_representation]
    df.to_csv(file, sep="\t",index=False)
```

```python
for file in glob(".\\spanish\\*.tsv"):
    df = pd.read_csv(file, sep="\t")
    counts = [3,3,3,3,2,2,3,3,2]
    pairs = ["english_word numeric", "numeric english_word", "spanish_word numeric", "numeric spanish_word", "english_word english_word", "spanish_word spanish_word", "english_word spanish_word","spanish_word english_word", "numeric numeric"]
    combos = np.random.choice([pair for sublist in [[x]*y for x,y in zip(pairs,counts)] for pair in sublist], size = 24, replace=False)
    df.equation_representation, df.comparison_representation = [x.split(" ")[0] for x in combos], [x.split(" ")[1] for x in combos]
    df.to_csv(file, sep="\t",index=False)
```

```python
# Change extensions of word files
work_dir = os.path.join(os.path.dirname(os.getcwd()),"stimuli","numerals"); os.chdir(work_dir)
[os.rename(x, x.replace("_w","_e")) for x in glob(".\\english\\*")]
[os.rename(x, x.replace("_w","_s")) for x in glob(".\\spanish\\*")]
```

```python
x = []
for file in glob(".\\spanish\\*.tsv"):
    df = pd.read_csv(file, sep = "\t")
    if list(df.equation_representation.value_counts()) + list(df.comparison_representation.value_counts()) != [8]*6:
        x.append(False)
    else: x.append(True)

all(x)
```
