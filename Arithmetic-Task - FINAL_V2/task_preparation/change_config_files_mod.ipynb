```python
#Replace word trials with english_word
for file in glob(".\\english\\*.tsv"):
    df = pd.read_csv(file, sep="\t")
    df.equation_representation = [x.replace("word", "english_word") if x == "word" else x for x in df.equation_representation]
    df.comparison_representation = [x.replace("word", "english_word") if x == "word" else x for x in df.comparison_representation]
    df.to_csv(file, sep="\t",index=False)
```

```python
#Modify training config files
import os, pandas as pd, numpy as np, shutil, copy, random; from glob import glob

#Assign work_dir
work_dir = os.path.join(os.path.dirname(os.getcwd()),"config"); os.chdir(work_dir)

files = zip(glob(".\\english\\*.tsv"),glob(".\\spanish\\*.tsv"))

feedback_1 = ["noninformative"]*4 + ["informative"]*20
feedback_2 = ["noninformative"]*7 + ["informative"]*17

for english_file, spanish_file in files:
    df_english, df_spanish = pd.read_csv(english_file, sep="\t"), pd.read_csv(spanish_file, sep="\t")
    #Drop feedback column
    df_english.drop(["feedback"], axis = 1, inplace=True); df_spanish.drop(["feedback"], axis = 1, inplace=True)
    
    #Randomize feedback_1 and feedback_2
    feedback_1 = np.random.choice(feedback_1, size = len(feedback_1), replace=False)
    feedback_2 = np.random.choice(feedback_2, size = len(feedback_2), replace=False)

    # Insert englis
    df_english.insert(5, "feedback_1", feedback_1)
    df_spanish.insert(5, "feedback_1", feedback_1)

    df_english.insert(6, "feedback_2", feedback_2)
    df_spanish.insert(6, "feedback_2", feedback_2)

    # Save files

    df_english.to_csv(english_file, sep="\t",index=False)
    df_spanish.to_csv(spanish_file, sep="\t",index=False)



```
